```python
input_num = int(input("Enter the values till you want to see the series: "))

def fib(z):
    a, b = 0, 1
    series = []
    for _ in range(z):
        series.append(a)
        a, b = b, a + b
    return series

# Print the series up to the specified number of terms
print(fib(input_num))



```
